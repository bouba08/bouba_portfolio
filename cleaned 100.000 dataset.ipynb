```python
df = pd.read_csv(
    "train.csv",
    header = 0,
    usecols =["ID","Age","Occupation","Annual_Income","Monthly_Inhand_Salary","Outstanding_Debt","Payment_Behaviour","Monthly_Balance","Credit_Score"],
   
)
df
```

```python

pd.set_option("display.float_format", lambda x: '%.2f' %x)
df

```

```python
df.dropna(inplace = True)
df.shape
    
```

```python
df["Payment_Behaviour"].unique()
```

```python
df["Payment_Behaviour"] = df["Payment_Behaviour"].replace('!@9#%8',' ')
df["Payment_Behaviour"].unique()

```

```python
df["Annual_Income"].unique
```

```python
df['Annual_Income'] = df['Annual_Income'].str.replace('[^\d.]', '', regex=True)
df['Annual_Income'] = df['Annual_Income'].astype(float)
df["Annual_Income"].unique()

```

```python
df.dropna(inplace = True)
df.shape
```

```python
df = df.reset_index(drop=True) 
```
